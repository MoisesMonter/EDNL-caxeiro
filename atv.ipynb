```python
import itertools

def bruteforce(graph):
    nodes = list(graph.nodes)
    min_cost = float('inf')
    optimal_path = None

    permutations = itertools.permutations(nodes[1:])
    print("Permutações possíveis:")
    for perm in permutations:
        path = [nodes[0]] + list(perm) + [nodes[0]]
        if all(graph.has_edge(path[i], path[i+1]) for i in range(len(path)-1)):
            cost = sum(graph[path[i]][path[i+1]]['weight'] for i in range(len(path)-1))
            if cost < min_cost:
                min_cost = cost
                optimal_path = path

    return optimal_path, min_cost


optimal_path_tsp, min_cost_tsp = bruteforce(G)
print("Caminho ótimo para o Problema do Caixeiro Viajante:", optimal_path_tsp)
print("Custo mínimo para o Problema do Caixeiro Viajante:", min_cost_tsp)
```
